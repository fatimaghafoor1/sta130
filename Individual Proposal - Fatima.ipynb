Research question 3:
How does age/getting older affect a persons life satisfaction? Does it increase or decrease with increasing age?


Variable(s) for analysis 3:
DEMO_age and WELLNESS_life_satisfaction
This analysis aims to create a simple linear regression to figure out if there is any relationship between these two variables.


Assumptions:
Assume constant variance of residuals (errors) across all ages.
Assume linear relationship between age and life satisfaction.


The code you could use for this is as follows:
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt  
data = pd.read_csv("CSCS_data_anon.csv")  
print(data.isnull().sum())
data = data.dropna(subset=['DEMO_age', 'WELLNESS_life_satisfaction'])
print(np.isinf(data['DEMO_age']).sum())
print(np.isinf(data['WELLNESS_life_satisfaction']).sum())
data = data[~np.isinf(data['DEMO_age'])]
data = data[~np.isinf(data['WELLNESS_life_satisfaction'])]
X = sm.add_constant(data['DEMO_age'])  
Y = data['WELLNESS_life_satisfaction']  
model = sm.OLS(Y, X).fit()
print(model.summary())
plt.figure(figsize=(10, 6))
plt.scatter(data['DEMO_age'], data['WELLNESS_life_satisfaction'], color='blue', label='Data Points')  # Scatter plot of the data
plt.xlabel('Age')
plt.ylabel('Wellness Life Satisfaction')
plt.title('Simple Linear Regression: Age vs. Wellness Life Satisfaction')
predicted_values = model.predict(X)
plt.plot(data['DEMO_age'], predicted_values, color='red', label='Regression Line')
plt.legend()
plt.show()

Possible result for analysis 3:

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt  
data = pd.read_csv("CSCS_data_anon.csv")  
print(data.isnull().sum())
data = data.dropna(subset=['DEMO_age', 'WELLNESS_life_satisfaction'])
print(np.isinf(data['DEMO_age']).sum())
print(np.isinf(data['WELLNESS_life_satisfaction']).sum())
data = data[~np.isinf(data['DEMO_age'])]
data = data[~np.isinf(data['WELLNESS_life_satisfaction'])]
X = sm.add_constant(data['DEMO_age'])  
Y = data['WELLNESS_life_satisfaction']  
model = sm.OLS(Y, X).fit()
print(model.summary())
plt.figure(figsize=(10, 6))
plt.scatter(data['DEMO_age'], data['WELLNESS_life_satisfaction'], color='blue', label='Data Points')  # Scatter plot of the data
plt.xlabel('Age')
plt.ylabel('Wellness Life Satisfaction')
plt.title('Simple Linear Regression: Age vs. Wellness Life Satisfaction')
predicted_values = model.predict(X)
plt.plot(data['DEMO_age'], predicted_values, color='red', label='Regression Line')
plt.legend()
plt.show()  
```
